```python
import yfinance as yf

# Download JPMorgan Chase (JPM) stock data for the last 1 year
stock = yf.download("JPM", period="1y")

# Display the first few rows
stock.head()
```

```python
print(stock.columns)
```

```python
print(stock.describe())
```

```python
import matplotlib.pyplot as plt

# Plot the Closing Price over time
plt.figure(figsize=(10, 5))
plt.plot(stock.index, stock["Close"], label="JPM Closing Price", color='blue')

# Adding title and labels
plt.title("JPMorgan Chase (JPM) Closing Price - Last 1 Year")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid()

# Show the plot
plt.show()

```
